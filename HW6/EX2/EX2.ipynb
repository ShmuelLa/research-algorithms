### Question 1 - Yearly Budget

```python
def education_budget(df, year: int) -> int:
    """
    Calculates education budget per year by filtering all
    rows where the income/expense flag is marked expense
    and the office category name is "Education"

    :param df: National budget dataframe
    :param year: The desired year to calculate
    :return: Education budget per the desired year
    """
    df = df[(df['שנה'] == year) 
        & (df['הוצאה/הכנסה'] == 'הוצאה') 
        & (df['שם סעיף'].str.contains('חינוך'))]
    return int(df['הוצאה נטו'].sum())
    
education_budget(df, 1997)
```

### Question 2 - Defense Budget Percentage

```python
def security_budget_ratio(df, year) -> float:
    """
    Calculates security budget percentage when compared to other offices.
    We sum all the expenses of the government and divide them by the
    expenses of the security budget as calculated in the former function

    :param df: National budget dataframe
    :param year: The desired year
    :return: Yearly security budget ratio
    """
    df = df[df['שנה'] == year]
    total_year_sum = df['הוצאה נטו'].sum()
    df = df[(df['הוצאה/הכנסה'] == 'הוצאה') 
        & (df['שם סעיף'] == 'משרד הבטחון')]
    return df['הוצאה נטו'].sum() / total_year_sum * 100

security_budget_ratio(df, 1997)
```

### Question 3 - Largest Yearly Budget

```python
def largest_budget_year(df, office: str) -> int:
    """
    Calculates largest office budget ever by name. First we filter out the
    chosen office by name, we iterate over the unique year values in the dataframe
    and sum each year expenses accordingly. At the end we return the
    maximum calculated value

    :param df: National budget dataframe
    :param office: Office name
    :return: Largest chosen office budget ever
    """
    df = df[df['שם סעיף'] == office]
    yearly_budget = {}
    for year in df['שנה'].unique():
        yearly_budget[year] = df[df['שנה'] == year]['הוצאה נטו'].sum()
    years = yearly_budget.values()
    return int(max(years))

largest_budget_year(df, 'משרד הבטחון')
```

### Question 4 - Average Salary Per Ministry

```python
def avg_office_salary_per_year(df, office, year) -> float:
    """
    Returns the average worker salary/expense for a specific 
    office in a specific year by calculating total salary expenses 
    in a certain year and dividing by the number of workers
    """
    df = df[(df['שם סעיף'] == office) 
            & (df['שם מיון רמה 1'] == 'שכר')
            & (df['שנה'] == year)]
    return round(df['הוצאה נטו'].sum() / df['שיא כח אדם'].max(), 3)

avg_office_salary_per_year(df, 'משרד הבטחון', 1997)
```
