# Exercise 1 - Intersection Plot

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve


def plotInterception(space, f, g):
    roots = np.unique([fsolve(lambda x : f(x) - g(x), val) for val in space])
    plt.plot(space, [f(i) for i in space], color='blue')
    plt.plot(space, [g(i) for i in space], color='orange')
    for root in roots:
        if round(f(root), 3) == round(g(root), 3):
            plt.plot(root, round(f(root), 3),"ro")
    plt.tight_layout()
    plt.grid()
    plt.show()
```

```python
plotInterception(np.linspace(-10,10,1000), lambda x: np.sin(x), lambda x: 0.2*x)
```

```python
plotInterception(np.linspace(-10,10,1000), lambda x: np.sin(x), lambda x: x**2)
```

```python
plotInterception(np.linspace(-10,10,1000), lambda x: x+1, lambda x: x**2)
```

```python
plotInterception(np.linspace(-10,10,1000), lambda x: x**3, lambda x: x**4)
```
